```python
import cv2
import os

photos_path = "./task3_photos/"
photos_dir = os.listdir(photos_path)


def nothing(x):
    pass


cv2.namedWindow("Blur_Filter")
cv2.createTrackbar("ksize_blur", "Blur_Filter", 1, 20, nothing)

cv2.namedWindow("Gaussain_Filter")
cv2.createTrackbar("ksize_gaussian", "Gaussain_Filter", 1, 20, nothing)
cv2.createTrackbar("sigmax", "Gaussain_Filter", 1, 20, nothing)

cv2.namedWindow("Median_Filter")
cv2.createTrackbar("ksize_median", "Median_Filter", 1, 20, nothing)

cv2.namedWindow("Bilateral_Filter")
cv2.createTrackbar("ksize_bilateral", "Bilateral_Filter", 1, 20, nothing)

i = 0
while i< len(photos_dir):
    img = cv2.imread(photos_path + photos_dir[i])
    img = cv2.resize(img, (400, 500))

    # Blur Filter:
    ksize_blur = cv2.getTrackbarPos("ksize_blur", "Blur_Filter")
    ksize_blur = ksize_blur if ksize_blur%2 == 1 else ksize_blur +1
    blur_filter = cv2.blur(img, ksize=(ksize_blur, ksize_blur))

    # Gaussian Filter:
    ksize_gaus = cv2.getTrackbarPos("ksize_gaussian", "Gaussain_Filter")
    ksize_gaus = ksize_gaus if ksize_gaus%2 == 1 else ksize_gaus +1
    sigmax = cv2.getTrackbarPos("sigmax", "Gaussain_Filter")
    gaussain_filter = cv2.GaussianBlur(
        img, ksize=(ksize_gaus, ksize_gaus), sigmaX=sigmax
    )

    # Median Filter:
    ksize_median = cv2.getTrackbarPos("ksize_median", "Median_Filter")
    ksize_median = ksize_median if ksize_median%2 == 1 else ksize_median +1
    median_filter = cv2.medianBlur(img, ksize=ksize_median)

    # Boilateral Filter:
    ksize_bil = cv2.getTrackbarPos("ksize_bilateral", "Bilateral_Filter")
    ksize_bil = ksize_bil if ksize_bil % 2 == 1 else ksize_bil +1
    bilateral_filter = cv2.bilateralFilter(img, ksize_bil, 75, 75)

    txt1 = "Press 'n' to continue to the next Image"
    txt2 = "Press 's' to save the modified Images"
    cv2.putText(img, txt1, (10, 15), cv2.FONT_HERSHEY_COMPLEX_SMALL, 0.7, (0, 255, 255))
    cv2.putText(img, txt2, (10, 35), cv2.FONT_HERSHEY_COMPLEX_SMALL, 0.7, (0, 255, 255))

    # Showing Images:
    cv2.imshow("Original Image", img)
    cv2.imshow("Blur_Filter", blur_filter)
    cv2.imshow("Gaussain_Filter", gaussain_filter)
    cv2.imshow("Median_Filter", median_filter)
    cv2.imshow("Bilateral_Filter", bilateral_filter)

    k = cv2.waitKey(1)
    
    if k == ord("n"):  ## Go to next photo
        i+=1

    elif k == ord("s"):
        cv2.imwrite(photos_path + photos_dir[i][0] + "_modified_blur_filter.jpg", blur_filter)
        cv2.imwrite(
            photos_path + photos_dir[i][0] + "_modified_gaussian_filter.jpg", gaussain_filter
        )
        cv2.imwrite(photos_path + photos_dir[i][0] + "_modified_median_filter.jpg", median_filter)
        cv2.imwrite(
            photos_path + photos_dir[i][0] + "_modified_bilateral_filter.jpg", bilateral_filter
        )
         
    elif k == 27:
        break

cv2.destroyAllWindows()
```
